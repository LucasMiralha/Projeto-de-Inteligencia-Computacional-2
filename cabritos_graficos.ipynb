```python
import pandas as pd
import matplotlib.pyplot as plt

# LER O CSV (seu arquivo com ;)
df = pd.read_csv("cabritos_fitness.csv", 
                sep=";",              # separador ;
                skipinitialspace=True # ignora espaços depois do ;
)

# Garantir que tudo está em formato numérico
cols_num = ["Generation", "BestFitness", "AvgFitness", "globalBestFitness", "WorstFitness"]
df[cols_num] = df[cols_num].apply(pd.to_numeric, errors="coerce")
```

```python
# Cria uma nova figura (janela de gráfico) com tamanho 8x5 polegadas
plt.figure(figsize=(8, 5))

# Plota a curva da melhor fitness por geração
plt.plot(df["Generation"], df["BestFitness"], label="BestFitness")

# Plota a curva da fitness média por geração
plt.plot(df["Generation"], df["AvgFitness"], label="AvgFitness")

# Plota a curva da pior fitness por geração
plt.plot(df["Generation"], df["WorstFitness"], label="WorstFitness")

# Define o rótulo (nome) do eixo X
plt.xlabel("Geração")

# Define o rótulo (nome) do eixo Y
plt.ylabel("Fitness")

# Define o título do gráfico
plt.title("Evolução da Fitness dos Cabritos (dados brutos)")

# Exibe a legenda usando os labels definidos em cada plt.plot
plt.legend()

# Ativa a grade no fundo do gráfico para facilitar a leitura
plt.grid(True)

# Salva o gráfico em um arquivo PNG com alta resolução (300 dpi)
# bbox_inches="tight" remove espaços em branco desnecessários nas bordas
plt.savefig("fitness_cabritos_bruto.png", dpi=300, bbox_inches="tight")

# Mostra o gráfico na tela (útil em notebooks e ambientes interativos)
plt.show()
```

```python
# Cria uma cópia do DataFrame original para não alterar os dados brutos
df_smooth = df.copy()

# Define o tamanho da janela da média móvel.
# Quanto maior o valor de 'window', mais suave (menos "serrilhado") fica o gráfico.
window = 20

# Calcula a média móvel da coluna BestFitness e salva em uma nova coluna 'Best_smooth'
df_smooth["Best_smooth"] = df["BestFitness"].rolling(window=window).mean()

# Calcula a média móvel da coluna AvgFitness e salva em 'Avg_smooth'
df_smooth["Avg_smooth"] = df["AvgFitness"].rolling(window=window).mean()

# Calcula a média móvel da coluna WorstFitness e salva em 'Worst_smooth'
df_smooth["Worst_smooth"] = df["WorstFitness"].rolling(window=window).mean()

# Cria uma nova figura (janela de gráfico) com tamanho 8x5 polegadas
plt.figure(figsize=(8, 5))

# Plota a curva suavizada da melhor fitness por geração
plt.plot(df_smooth["Generation"], df_smooth["Best_smooth"], label="BestFitness (média móvel)")

# Plota a curva suavizada da fitness média por geração
plt.plot(df_smooth["Generation"], df_smooth["Avg_smooth"], label="AvgFitness (média móvel)")

# Plota a curva suavizada da pior fitness por geração
plt.plot(df_smooth["Generation"], df_smooth["Worst_smooth"], label="WorstFitness (média móvel)")

# Define o rótulo (nome) do eixo X
plt.xlabel("Geração")

# Define o rótulo (nome) do eixo Y
plt.ylabel("Fitness")

# Define o título do gráfico
plt.title("Evolução da Fitness dos Cabritos (suavizada)")

# Exibe a legenda usando os labels definidos em cada plt.plot
plt.legend()

# Ativa a grade no fundo do gráfico para facilitar a leitura
plt.grid(True)

# Salva o gráfico suavizado em um arquivo PNG com alta resolução (300 dpi)
# bbox_inches="tight" remove espaços em branco desnecessários nas bordas
plt.savefig("fitness_cabritos_suavizado.png", dpi=300, bbox_inches="tight")

# Mostra o gráfico na tela (útil em notebooks e ambientes interativos)
plt.show()
```

```python
# Cria uma nova figura (janela de gráfico) com tamanho 8x5 polegadas
plt.figure(figsize=(8, 5))

# Plota a curva do Global Best Fitness ao longo das gerações.
# 'Generation' é o eixo X (tempo/iteração do algoritmo)
# 'globalBestFitness' é o melhor valor de fitness encontrado até aquela geração.
plt.plot(df["Generation"], df["globalBestFitness"])

# Define o rótulo (nome) do eixo X
plt.xlabel("Geração")

# Define o rótulo (nome) do eixo Y
plt.ylabel("Global Best Fitness")

# Define o título do gráfico
plt.title("Convergência do Global Best Fitness - Cabritos vs Pepinos")

# Ativa a grade no fundo do gráfico para facilitar a leitura
plt.grid(True)

# Salva o gráfico em um arquivo PNG com alta resolução (300 dpi)
# bbox_inches="tight" remove espaços em branco desnecessários nas bordas
plt.savefig("global_best_cabritos.png", dpi=300, bbox_inches="tight")

# Mostra o gráfico na tela (útil em notebooks e ambientes interativos)
plt.show()
```
